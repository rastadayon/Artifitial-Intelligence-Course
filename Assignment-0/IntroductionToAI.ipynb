### Chosing the variable with the largest correlation with Chance of Admit variable

To find which of the variablesin the dataframe have the largest linear correlation with Chance of Admit variable we can have a good guess by plotting them however to make sure, we calculate the correlation value. Correlation is a statistical measure of linear correlation between two variable. There are a few different mothods for calculating the correlation, one of them is the Pearson correlation coefficient. The Pearson correlation coefficient is calculated with this formula:

\begin{align*}
\rho_{X,Y} = \frac{cov(X,Y)}{\sigma_X \sigma_Y}
\end{align*}

In which the $cov(X,Y)$ is the covariance of the two variables.
The $\sigma_X$ and $\sigma_y$ are standard deviations of X and Y variables and $\rho_{X,Y}$ represents the correlation value.

The coefficient value falls somewhere between -1 and +1 where a +1 indicates a perfect positive relationship and -1 indicates a perfect negative relationship.

Pandas library has a corr() function which when called without arguments on a data frame calculates the pairwise correlation of the columns (excluding NaN values) and returns a data frame.

The returned data frame from corr function will have 1 along the main diagonals and it is also symmetric.

```python
data.corr()
```

Every variable has a correlation of 1 with itself(obviously). Variable CGPA has the largest correlation with variable Chance of Admit with the value of 0.855675 which was also observable in the plots.

It is a given that correlation value is a good representation of how two variables are linearly related, therefor we can observe that CGPA is the most linearly related varible to variable Chance of Admit.

## One variable linear regression

In this part we want to find a good linear estimation for the Chance of Admit variable with the input value of CGPA that was found the most linearly related to the Chance of Admit variable in part 2.

For simplicity we are going to use the derivation method in this project however there are many other good methods such as Gradient Descent or Stochastic Gradient Descent and so on. About the method used for this project there will be futher detailed explanation.

### Hypothesis function

This function given the theta values and input variables (X) produces an estimation for the output.

The output will be like the following:
\begin{align*}
h_\theta(x) = \theta_0 + \theta_1 x
\end{align*}
$\theta_0$ and $\theta_1$ are given as inputs to the function. $\theta_1$ is the slope for the linear relation and $\theta_0$ is the y-intercept.

```python
def hypothesisFunc(theta0, theta1, X):
    return X* theta1 + theta0
```

When calling this method, the X input is a vector so when multiplied by theta1 every element will be multiplied and when added with theta0 the same applies. The theta values could have also been vectors however for this problem they were considered as numeric values.

### Cost function

To check if the estimation is a good one we should find how far away it is from the actual value. To determine this cost we use a cost function. This cost function calculates **Mean Square Error** which is obtained from the following equation:
\begin{align*}
J_(\theta_0,\theta_1) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)} - y^{(i)})^2
\end{align*}



```python
def costFunc(theta0, theta1 , X ,y ):
    m = len(y)
    
    predictions = hypothesisFunc(theta0,theta1,X)
    cost = ((predictions-y)**2).sum()/(2*m)
    return cost
```

X is the vector of input values (CGPA scores) and y is the target value (Chance of Admit), theta0 and theta1 are same as the last part. Also m is the length of y vector (same as the number of train datas).

### Generating theta values 

The method used to find the best values for $\theta_0$ and $\theta_1$ here is to find the values that make the cost the least possible value. For this porpuse what we need is to do two partial derivations one for each variable. From the two resaulted equations the 2 variables can be obtained. 

These are a summary of the mathematics used for calculating theta0 and theta1:

$$ SSE = \sum\limits_{i=1}^{n} {(Y_i - \hat Y_i)^2} $$

$$ 
SSE = 
n( \bar Y - \theta_0 - \theta_1 \bar X)^2 
+ \left ( 
\theta_1 - 
{ \sum\limits_{i=1}^{n} { (X_i - \bar X)(Y_i - \bar Y) } 
\over 
\sum\limits_{i=1}^{n} {(X_i - \bar X)} } \right ) ^2
+
\sum\limits_{i=1}^{n} {(Y_i - \bar Y)^2}
\left ( 1 - \left [ { \sum\limits_{i=1}^{n} {(X_i - \bar X)(Y_i - \bar Y)}
\over 
{ \sqrt { \sum\limits_{i=1}^{n} {(X_i - \bar X)^2} \sum\limits_{i=1}^{n} {(Y_i - \bar Y)^2} } } } \right ] ^ 2 
\right )
$$

$$
n ( \bar Y - \theta_0 - \theta_1 \bar X ) ^ 2 = 0
$$


$$
\left ( 
\theta_1 - 
{ \sum\limits_{i=1}^{n} {(X_i - \bar X)(Y_i - \bar Y) }
\over
\sum\limits_{i=1}^{n} {(X_i - \bar X)^2 } } 
\right ) ^ 2
= 0
$$


$$
\hat \theta_1 = 
{ \sum\limits_{i=1}^{n} {(X_i - \bar X)(Y_i - \bar Y) }
\over
\sum\limits_{i=1}^{n} {(X_i - \bar X)^2 } } 
$$

$$
\hat \theta_0 =
\bar Y - \hat \theta_1 \bar X
$$

The specific mathematics of how these two variables are calculated can be found in this [link](https://www.amherst.edu/system/files/media/1287/SLR_Leastsquares.pdf)

```python
def estimationFunc(X, y):
    theta1 = ((X - X.mean())*(y-y.mean())).sum() / ((X-X.mean())**2).sum()
    theta0 = y.mean() - theta1*(X.mean())
    return(theta0, theta1)
```

This function inputs X and y vectors and outputs the theta values. The theta values are obtained from the equations that were talked about in the previous sections. Since in the Chance of Admit some values are NaN and we want to only work with the valid values we can drop the unwanted values (same way described in the previous parts) and create a new data frame called traindf.

```python
traindf = newdf.dropna()
```

X and y vectors (series) are validated in the following line of code.

```python
X = traindf['CGPA']
y = traindf['Chance of Admit']
```

```python
theta0, theta1 = estimationFunc(X,y)
theta0, theta1
```
